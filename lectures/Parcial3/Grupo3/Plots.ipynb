```python
import numpy as np
import matplotlib.pyplot as plt

def plot(data, a,b,c,d, style):
    fig = plt.figure(figsize = (6,6))
    im = plt.imshow(data , cmap = style, extent=[a,b,c,d])

    plt.xlabel("x")
    plt.ylabel("y")
    plt.colorbar(im);
```

```python
data = np.loadtxt("Example.txt")
plot(data,0,2,0,1, 'jet')
plt.savefig('Example.png')
```

```python
data = np.loadtxt("Dipole.txt")
plot(data, -2,2,-2,2, 'seismic')
plt.savefig('Dipole.png')
```

```python
data = np.loadtxt("PositiveCharges.txt")
plot(data, -2,2,-2,2,'Reds')
plt.savefig('PositiveCharges.png')
```

```python
data = np.loadtxt("Excercise.txt") 
plot(data,0,6,0,5,'jet')
plt.savefig('Excercise.png')
```

```python
data = np.loadtxt("Stagnation.txt")
plot(data,-2,2,0,4,'jet')
plt.savefig('Stagnation.png')
```
